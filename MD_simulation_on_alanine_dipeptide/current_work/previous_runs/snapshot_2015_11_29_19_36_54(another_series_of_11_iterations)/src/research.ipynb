```python
# fix autoencoder, compare different datasets
%matplotlib inline

folder_list = ['../target/network_%d/' % item for item in range(1, 12)]
a=pickle.load(open('../resources/network_1.pkl','rb'))
my_file_list = ['../target/unbiased/unbiased_output_coordinates.txt']

i = 1

for item in folder_list:
    
    b=plotting(a)
    my_file_list += coordinates_data_files([item])._list_of_coor_data_files
    color = [item[2] for item in a.get_many_dihedrals_from_coordinates_in_file(my_file_list)]
    temp_fig = b.plotting_with_coloring_option(plotting_space = "phipsi",
                                        list_of_coordinate_files_for_plotting=my_file_list, color_option='other', 
                                        other_coloring = color,
                                        axis_ranges=[[-4,4],[-4,4]],
                                        title="plotting in phi-psi space in iteration %d" % i
                                        )
#     temp_fig.savefig('iteration_%d.png' % i)
    
    i += 1
plt.show()
```
